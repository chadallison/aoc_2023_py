```python
import numpy as np
import pandas as pd
import re
```

```python
with open("input.txt", "r") as f:
    lines = [x.strip() for x in f.readlines()]

df = pd.DataFrame({"line": lines})

df[["winning", "have"]] = df["line"].str.extract(".*: (.*) \\| (.*)")
df = df.drop("line", axis = 1)

df["overlap"] = df.apply(lambda row: list(set(row["winning"]).intersection(row["have"])), axis = 1)

df
```
